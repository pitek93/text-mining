```python
# reading in words from all txt files:
import glob
files = glob.glob("*.txt")

words = []
for file in files:
    f = open(file, encoding='utf8')
    for line in f:
        if line.startswith("*** START OF"):
            break
    for line in f:
        if line.startswith("*** END OF"):
            break
        words.extend(line.strip().split(' '))

# removing empty strings
words = [w for w in words if w != '']

N = 15
print(f"first {N} out of {len(words)} words:\n{words[:N]}")
```

```python
# randomly shuffling all words
import random
random.shuffle(words)
print('shuffled')
```

```python
# distinct number of words as a functin of words number
import numpy as np

words_num = list(map(int, np.linspace(1, len(words), 200)))
distinct_words_num = [len(set(words[:wn])) for wn in words_num]

print(f"number of words:                    {words_num[:10]}")
print(f"corresponding number of distinct words:{distinct_words_num[:10]}")
```

```python
"number of words in a function of distinct words"
plt.plot(words_num, distinct_words_num)
plt.xlabel('number of words')
plt.ylabel('number of distinct words', rotation=80)
plt.gcf().axes[0].xaxis.get_major_formatter().set_scientific(False)
no_verbose = plt.xticks(rotation=50)
```

```python
# Heaps law:
from math import log10
words_num_log = [math.log10(x) for x in words_num]
distinct_words_num_log = [math.log10(x) for x in distinct_words_num]

import matplotlib.pyplot as plt
plt.plot(words_num_log, distinct_words_num_log)

from sklearn.linear_model import LinearRegression
# fitting linear estimator
X = np.array(words_num_log).reshape(-1, 1)
y = np.array(distinct_words_num_log).reshape(-1, 1)
reg = LinearRegression().fit(X, y)
coef = reg.coef_[0][0]
intercept = reg.intercept_[0]

axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = intercept + coef * x_vals
plt.plot(x_vals, y_vals, '--', label='estimate')
plt.xlabel('log10T')
plt.ylabel('log10M')

print(coef, intercept)
```

```python
def M(k, T, b):
    return k*(T)**b

# k=(30 ... 100) eg 44
# b=(0.4 ... 0.6) eg 0.49
k = 10**intercept
b = coef
print(f"Heaps' parameters: k={k:.2f}, b={b:.2f}")

observed = len(set(words[:800000]))
estimate = M(k, 800000, b)
print("observed    estimate")
print(f"{observed:.2f}    {estimate:.2f}")

#log10(M) = b*log10(T) + log10(k)
```

```python
from collections import Counter

def word_count_dict(filename):
    f = open(filename, encoding='utf8')
    c = Counter()
    
    for line in f:
        if line.startswith("*** START OF"):
            break
            
    for line in f:
        if line.startswith("*** END OF"):
            break
            
        c.update(line.strip().split(' '))

    return c

files = glob.glob("*.txt")
counters = [word_count_dict(filename) for filename in files]
concatenated_counters = sum(counters, Counter())
```

```python
sum(concatenated_counters.values())
```

```python
rank = [i for i in range(1, len(concatenated_counters))]
```

```python
most_common = concatenated_counters.most_common()
rank = [i for i in range(1, len(concatenated_counters)+1)]
values = [i[1] for i in most_common]
```

```python
rank_log = [log10(i) for i in rank]
values_log = [log10(i) for i in values]
```

```python
plt.plot(rank_log, values_log)
from sklearn.linear_model import LinearRegression
# fitting linear estimator
X = np.array(rank_log).reshape(-1, 1)
y = np.array(values_log).reshape(-1, 1)
reg = LinearRegression().fit(X, y)
coef = reg.coef_[0][0]
intercept = reg.intercept_[0]

axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = intercept + coef * x_vals
plt.plot(x_vals, y_vals, '--', label='estimate')
plt.xlabel('log10T')
plt.ylabel('log10M')

print(coef, intercept)
```
